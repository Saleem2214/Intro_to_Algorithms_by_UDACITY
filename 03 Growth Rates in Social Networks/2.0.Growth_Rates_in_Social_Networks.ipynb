```python
# 3.4 Ring Network ارتباط الحلقة

    # Nodes = (Vertices) = n
    # Edges = (Links) = m
# Ring Network: If n = 20 >>> m = 20

def make_link(G, node1, node2):
    if node1 not in G:
        G[node1] = {}
    (G[node1])[node2] = 1
    if node2 not in G:
        G[node2] = {}
    (G[node2])[node1] = 1
    return G

# Make an empty graph
a_ring = {}
n = 5

# Add in the edges
for i in range(n):
    make_link(a_ring, i, (i+1)%n)
    
# How many nodes?
print("Nodes (Vertices) =",len(a_ring))
    
# How many edges?
print("Edges (Links) =",sum([len(a_ring[node]) for node in a_ring.keys()])/2)
print("A ring:",a_ring)
```

```python
# 3.6 Big-Theta (Θ) ثيتا الكبيرة

    # NO. of Edges = 2n-2Squ(n) ∈ Θ(n)
    # Θ(g(n))
    # f(n) ∈ Θ(g(n)) 
    #We can take a function like ½n² and just think of it as Θ(n²).

import matplotlib.pyplot as plt
import numpy as np

# تعريف الدوال والثوابت
n = np.linspace(1, 10, 400)  # مدى قيم n
f_n = n**2 + 3*n + 2          # مثال للدالة f(n)
g_n = n**2                     # مثال للدالة g(n)

c1 = 1.0                       # ثابت أقل حد
c2 = 2.0                       # ثابت أعلى حد
n0 = 4                         # نقطة البدء n₀

# إنشاء الرسم
plt.figure(figsize=(10, 6))
plt.plot(n, f_n, label=r'$f(n) = n^2 + 3n + 2$', color='blue')
plt.plot(n, c1 * g_n, '--', label=r'$c_1 \cdot g(n)$', color='red')
plt.plot(n, c2 * g_n, '--', label=r'$c_2 \cdot g(n)$', color='green')

# تظليل المنطقة بين الحدين
plt.fill_between(n, c1 * g_n, c2 * g_n, where=(n >= n0), color='yellow', alpha=0.3)

# إضافة العلامات والتنسيق
plt.title(r'Big Theta Notation: $f(n) \in \Theta(g(n))$', fontsize=14)
plt.xlabel(r'$n$', fontsize=12)
plt.ylabel(r'$T(n)$', fontsize=12)
plt.axvline(n0, color='black', linestyle=':', label=r'$n_0$')
plt.annotate(r'$\forall n > n_0$', xy=(n0+0.5, 50), fontsize=12)
plt.legend()
plt.grid(True)

plt.show() # إظهار الرسم
```

```python
# 3.8 Big-Theta Examples امثلة ثيتا الكبير

import matplotlib.pyplot as plt
import numpy as np

# تعريف نطاق n
n = np.linspace(1, 20, 400)  # من 1 إلى 20 بدقة عالية

# حساب الدوال
f = 2 * n - 2 * np.sqrt(n)
c1 = 1.0  # الحد الأدنى
c2 = 2.0  # الحد الأعلى
lower_bound = c1 * n
upper_bound = c2 * n

# تحديد n0 (نقطة بدء التحقق من الثوابت)
n0 = 4

# إنشاء الرسم
plt.figure(figsize=(10, 6))
plt.plot(n, f, label=r'$2n - 2\sqrt{n}$', color='purple', linewidth=2)
plt.plot(n, lower_bound, label=rf'$c_1 n = {c1}n$', linestyle=':', color='orange')
plt.plot(n, upper_bound, label=rf'$c_2 n = {c2}n$', linestyle='--', color='red')

# تظليل المنطقة بين الحدود بعد n0
plt.fill_between(
    n[n >= n0], 
    lower_bound[n >= n0], 
    upper_bound[n >= n0], 
    color='lightgreen', 
    alpha=0.3, 
    label=r'$\Theta(n)$ لـ $n \geq 4$'
)

# خط عمودي للإشارة إلى n0
plt.axvline(n0, color='gray', linestyle='--', alpha=0.7, label=r'$n_0 = 4$')

# تنسيق الرسم
plt.xlabel('n', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.title(r'Verification: $2n - 2\sqrt{n} \in \Theta(n)$', fontsize=14)
plt.legend(loc='upper left', fontsize=10)
plt.grid(True, linestyle=':', alpha=0.5)
plt.ylim(0, 40)  # ضبط نطاق المحور الرأسي

plt.show() # عرض الرسم
```

```python
# 3.9 Other sets of Functions مجموعات أخرى من الوظائف

    # F(n) ∈ o(g(n)) ⇒ F(n) < g(n)
    # F(n) ∈ O(g(n)) ⇒ F(n) ≤ g(n)
    # F(n) ∈ Θ(g(n)) ⇒ F(n) = g(n)
    # F(n) ∈ Ω(g(n)) ⇒ F(n) ≥ g(n)
    # F(n) ∈ ω(g(n)) ⇒ F(n) > g(n)

import matplotlib.pyplot as plt
import numpy as np

# تعريف نطاق n
n = np.linspace(1, 20, 400)

# تعريف الدوال الأساسية
g_n = n  # الدالة المرجعية g(n) = n

# اختيار دوال تمثل كل رمز بدقة
f_o = np.log(n)     # o(g(n)): نمو أبطأ بكثير (مثال: log(n))
f_O = 2 * n + 3     # O(g(n)): نمو لا يتجاوز g(n)
f_Theta = n         # Θ(g(n)): نمو مكافئ لـ g(n)
f_Omega = 0.5 * n   # Ω(g(n)): نمو على الأقل مثل g(n)
f_omega = n ** 1.5  # ω(g(n)): نمو أسرع بكثير (مثال: n^1.5)

# إنشاء الرسم
plt.figure(figsize=(14, 8))

# رسم الدوال مع تخصيص الأنماط
plt.plot(n, g_n, label=r'$g(n) = n$', color='black', linewidth=3, linestyle='-')
plt.plot(n, f_o, label=r'$f(n) \in o(g(n)) \ (\log(n))$', color='red', linestyle='--')
plt.plot(n, f_O, label=r'$f(n) \in O(g(n)) \ (2n + 3)$', color='blue', linestyle='-.')
plt.plot(n, f_Theta, label=r'$f(n) \in \Theta(g(n)) \ (n)$', color='green', linestyle='-')
plt.plot(n, f_Omega, label=r'$f(n) \in \Omega(g(n)) \ (0.5n)$', color='purple', linestyle=':')
plt.plot(n, f_omega, label=r'$f(n) \in \omega(g(n)) \ (n^{1.5})$', color='orange', linestyle='--')

# تنسيق الرسم
plt.xlabel('n', fontsize=16)
plt.ylabel('Value', fontsize=16)
plt.title('العلاقات بين الرموز الكبيرة حسب التعريفات القياسية', fontsize=14)
plt.legend(loc='upper left', fontsize=14)
plt.grid(True, linestyle=':', alpha=0.7)
plt.ylim(0, 100)  # ضبط نطاق المحور الرأسي

plt.show() # عرض الرسم
```
