```python
# 16.5 Names And Boxes Problem مشكلة الأسماء والصناديق

    # https://en.wikipedia.org/wiki/100_prisoners_problem (go the Web).

    # (Problem):
    # there are 100 prisoners,
    # The warden will be give their freedom,
    # if they can win the follwing game:
    # 100 boxes on the table,
    # in each box will put on of prisoner's name,
    # each prisoner looks at only 50 boxes (50%) to find his name.
    # What algorithm will each prisoner follow?
    # 
    # (Solution):
    # 1. Each prisoner first opens the drawer labeled with their own number.
    # 2. If this drawer contains their number, they are done and were successful.
    # 3. Otherwise, the drawer contains the number of another prisoner, and they next open the drawer labeled with this number.
    # 4. The prisoner repeats steps 2 and 3 until they find their own number, or fail because the number is not found in the first fifty opened drawers.

import random
import matplotlib.pyplot as plt
import networkx as nx

def simulate_prisoners(num_prisoners=100, max_attempts=50):
    # إنشاء صناديق عشوائية
    boxes = list(range(1, num_prisoners+1))
    random.shuffle(boxes)
    
    # محاكاة إستراتيجية الدورة
    for prisoner in range(1, num_prisoners+1):
        current_box = prisoner
        found = False
        for _ in range(max_attempts):
            if boxes[current_box-1] == prisoner:
                found = True
                break
            current_box = boxes[current_box-1]
        if not found:
            return False
    return True

# محاكاة 10000 تجربة لحساب احتمالية النجاح
success = 0
trials = 10000
for _ in range(trials):
    if simulate_prisoners():
        success += 1

print(f"Success probability: {success/trials:.2%}")  # ~31%

# رسم دورة مثال توضيحي (مصغر)
def draw_sample_cycle():
    G = nx.DiGraph()
    sample_boxes = [2, 3, 4, 1, 5, 6, 7, 8]  # دورة بطول 4: 1→2→3→4→1
    for i, num in enumerate(sample_boxes, 1):
        G.add_edge(i, num)
        
    pos = nx.circular_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', 
            arrowsize=20, node_size=800)
    plt.title("Example permutation cycle (length 4)")
    plt.show()

draw_sample_cycle()
```

```python
# 16.11 Networks الشبكات

import networkx as nx
import matplotlib.pyplot as plt

# إنشاء شبكة نجمة بـ 5 عقد (1 مركز + 4 طرفية)
G = nx.star_graph(8)  # الرقم 4 يمثل عدد العقد الطرفية

# تسمية العقد
labels = {0: "Hub (Central)"}
for i in range(1, 9):
    labels[i] = f"Node {i}"

# رسم الشبكة
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, labels=labels, node_size=800, node_color='lightblue')
plt.title("Star Network with Shortest Paths through Hub")
plt.show()

# Go to Quiz (How many hubs ... efficiently?)
    # [1]
```
